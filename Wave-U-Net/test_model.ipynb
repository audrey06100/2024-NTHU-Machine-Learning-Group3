### waveunet_valid

```python
import os
import librosa
import numpy as np
from pesq import pesq
from pystoi import stoi

# 資料夾路徑
clean_dir = "/content/drive/MyDrive/audio/clean_valid"  # 放乾淨語音的資料夾
noisy_dir = "/content/drive/MyDrive/audio/noisy_valid"  # 放含噪語音的資料夾
enhanced_dir = "/content/drive/MyDrive/audio/WaveUNet_valid"  # 放增強後語音的資料夾

# 音檔讀取函數
def load_audio_files(directory, sample_rate=16000):
    audio_files = []
    file_list = sorted(os.listdir(directory))  # 確保文件順序一致
    for i, file in enumerate(file_list):
        filepath = os.path.join(directory, file)
        audio, _ = librosa.load(filepath, sr=sample_rate)
        audio_files.append(audio)
        if i % 100 == 0:
            print(f"Processed {i} files...")
    return audio_files

# 載入數據
clean_speech = load_audio_files(clean_dir)
print("clean_speech", len(clean_speech))
#noisy_speech = load_audio_files(noisy_dir)  # 如果需要也可以使用
#print("noisy_speech", len(noisy_speech))
enhanced_speech = load_audio_files(enhanced_dir)
print("enhanced_speech", len(enhanced_speech))

# 計算指標
pesq_scores = []
stoi_scores = []
nmse_scores = []

for clean, enhanced in zip(clean_speech, enhanced_speech):
    # PESQ 計算 (範例使用窄帶 nb，根據需要可改為 wideband wb)
    pesq_scores.append(pesq(16000, clean, enhanced, 'wb'))

    # STOI 計算
    stoi_scores.append(stoi(clean, enhanced, 16000, extended=False))

    # NMSE 計算
    mse = np.mean((clean - enhanced) ** 2)
    nmse = mse / np.mean(clean ** 2)
    nmse_scores.append(nmse)

# 計算平均值和標準差
pesq_mean, pesq_std = np.mean(pesq_scores), np.std(pesq_scores)
stoi_mean, stoi_std = np.mean(stoi_scores), np.std(stoi_scores)
nmse_mean, nmse_std = np.mean(nmse_scores), np.std(nmse_scores)

# 輸出結果
print(f"PESQ: Mean={pesq_mean:.2f}, Std={pesq_std:.2f}")
print(f"STOI: Mean={stoi_mean:.2f}, Std={stoi_std:.2f}")
print(f"NMSE: Mean={nmse_mean:.2f}, Std={nmse_std:.2f}")
```

PESQ: Mean=2.09, Std=0.57  
STOI: Mean=0.89, Std=0.10  
NMSE: Mean=0.05, Std=0.07
